```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.topendsports.com/events/worldcupsoccer/winners.htm"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python
title=soup.find('title')
print(title)
```

```python
print(title.text)
```

```python
table = soup.find('table')
print(table)
```

```python
Headings = table.find_all('th')
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
import pandas as pd
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
print(df.columns)
```

```python
ind=0
for i in rows[1:21]:
  row_data = i.find_all('td')
  every_row_data = [i.text for i in row_data]
  print(every_row_data)
  df.loc[ind] = every_row_data
  ind = ind+1
```

```python
df.to_csv("fifa winner list.csv")
```
